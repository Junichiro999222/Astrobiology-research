```python
import matplotlib as mpl
#mpl.use('Cairo') #for png, ps, pdf, svg, ...
mpl.use('Agg') #for png
import numpy as np
import matplotlib.pyplot as plt
from sympy import *
import math
from matplotlib import cm
from mpl_toolkits.axes_grid1 import make_axes_locatable
from scipy.optimize import curve_fit
import os
import csv
import pandas as pd
%matplotlib inline
import matplotlib.ticker as ticker

fig = plt.figure(figsize=(5,6))
ax1 = fig.add_subplot(111)

volume = np.arange(0,7.5,0.1)
volume2 = np.arange(5,7.5,0.001)


def find_quadratic_coefficients(axis, point1, point2):
    p = axis
    x1, y1 = point1
    x2, y2 = point2
    A = np.array([[x1**2, x1, 1],[x2**2, x2, 1],[p**2, p, 1]])
    B = np.array([y1, y2, 0])
    coefficients = np.linalg.solve(A, B)
    return coefficients

def quadratic_from_vertex_and_point(vertex, point):
    h, k = vertex
    x1, y1 = point
    a = (y1 - k) / ((x1 - h) ** 2)
    return a, h, k

def line_equation(point1, point2):
    x1, y1 = point1
    x2, y2 = point2
    m = (y2 - y1) / (x2 - x1)
    b = y1 - m * x1
    return m, b
def f_Hugoniot(a,b,c,v):
    Hugoniot = a*v**2 + b*v + c
    return Hugoniot

def find_intersection_vertex_form(a, h, k, m, b):
    A = a
    B = -m
    C = a * h**2 - 2 * a * h * h + k - b
    x_solutions = np.roots([A, B, C])
    y_solutions = a * (x_solutions - h)**2 + k
    return x_solutions, y_solutions

def f_curves(point1,point2,axis1,vertex):
    H_a,H_b,H_c = find_quadratic_coefficients(axis1,point1,point2)
    H_point1 = (5, f_Hugoniot(H_a,H_b,H_c,5))
    H_point2 = (6, f_Hugoniot(H_a,H_b,H_c,6))
    a, h, k  = quadratic_from_vertex_and_point(vertex, H_point1)
    isentrope = a*(volume2-h)**2 + k
    m, b = line_equation(H_point1, H_point2)
    slope =  m*volume2 + b
    Hugoniot = f_Hugoniot(H_a,H_b,H_c,volume)
    x_solutions, y_solutions = find_intersection_vertex_form(a, h, k, m, b)
    x_solutions[1]=5.84
    # print(x_solutions)
    for i in np.arange(0,len(volume2),1):
      if volume2[i] <= x_solutions[1] and volume2[i+1] >= x_solutions[1]:
        volume3 = np.arange(5,volume2[i],0.001)
        volume4 = np.arange(volume2[i],7.5,0.001)
    isentrope3 = a*(volume3-h)**2 + k
    isentrope4 = a*(volume4-h)**2 + k
    slope3 =  m*volume3 + b
    slope4 =  m*volume4 + b
    # print("Hugoniot = (v-{})^2  -{}".format(x1, y1))
    # print(Hugoniot)
    # print(isentrope)
    # print(slope)

    return isentrope,slope,Hugoniot,x_solutions, y_solutions,isentrope3,isentrope4,slope3,slope4,volume3,volume4

point1 = [5,3]
point2 = [6,0]

axis1 = 6.5
vertex = [7,-0.8]
isentrope,slope,Hugoniot,x_solutions, y_solutions,isentrope3,isentrope4,slope3,slope4,volume3,volume4 = f_curves(point1,point2,axis1,vertex)

pos_x = [5,6]
ticks_x = ['V_{shock}$', '$V_{initial}$']
pos_y = [0, 3]
ticks_y = ['$P_{initial}$', 'P_{shock}']
# 軸の zorder を設定
ax1.spines['bottom'].set_zorder(5)
ax1.spines['left'].set_zorder(5)
ax1.spines['top'].set_zorder(5)
ax1.spines['right'].set_zorder(5)

ax1.plot(volume,Hugoniot,'red',linewidth=1, label = "Hugoniot curve")
ax1.plot(volume2,isentrope,'blue',linewidth=1, label = "Isentrope")
ax1.plot(volume2,slope, color = "green", lw=1, linestyle = "dashed", label = "Rayleigh line")

ax1.scatter(6,0, color='blue',zorder=10,label ="Initial state",clip_on=False)
ax1.scatter(5,3, color='red',zorder=5,label ="Shock state")
# ax1.scatter(5,0, color='green',zorder=10)

ax1.set_xlabel('Specific volume',fontsize=15)
ax1.set_ylabel('Pressure',fontsize=15)
ax1.set_xlim([4.8,6.3])
ax1.set_ylim([0,3.5])

ax1.axhline(3, color = "black", alpha = 0.2, lw=1, linestyle = "dashed")
ax1.axvline(5, color = "black", alpha = 0.2, lw=1, linestyle = "dashed")
ax1.axvline(6, color = "black", alpha = 0.2, lw=1, linestyle = "dashed")


ax1.fill_between(volume3, isentrope3, where=( 0<= isentrope3 ), color='orange', alpha=0.2,edgecolor='none')
ax1.fill_between(volume4, slope4, where=( 0<= slope4), color='orange', alpha=0.2,edgecolor='none')
ax1.fill_between(volume3, isentrope3, slope3, where=( isentrope3 <= slope3 ), color='skyblue', alpha=0.2)

ax1.tick_params(labelsize=15)
ax1.set_xticks(pos_x)
ax1.set_xticklabels(ticks_x)
ax1.set_yticks(pos_y)
ax1.set_yticklabels(ticks_y)
ax1.xaxis.set_label_coords(0.5, -0.1)
ax1.yaxis.set_label_coords(-0.1, 0.5)
ax1.legend(fontsize=13,loc=1)
```
